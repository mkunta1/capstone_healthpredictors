```python
import pandas as pd
patients1 = pd.read_csv('C:\\Users\\Mahi2\\capstone\\capstone_healthpredictors\\data\\patients.csv')
# View a few rows (table)
print(patients1.head().to_string())
print("\nShape (rows, columns):", patients1.shape)
print(patients1.columns.tolist())
```

```python
cols_to_drop = ['PASSPORT', 'SSN', 'DRIVERS', 'MAIDEN', 'PREFIX', 'SUFFIX']
patients1.drop(columns=cols_to_drop, inplace=True)
```

```python
patients1['memberid'] = patients1['Id'].str[:8]
print("\nShape (rows, columns):", patients1.shape)
print(patients1.columns.tolist())
```

```python
unique_memberids = patients1['memberid'].nunique()
print(unique_memberids)

patients1.dtypes
```

```python
patients1['BIRTHDATE'] = pd.to_datetime(patients1['BIRTHDATE'], errors='coerce')
patients1['DEATHDATE'] = pd.to_datetime(patients1['DEATHDATE'], errors='coerce')
patients1['ZIP'] = patients1['ZIP'].astype('Int64').astype(str)
today = pd.Timestamp('today')

patients1['DEATHDATE'] = patients1['DEATHDATE'].fillna(today)

patients1['AGE'] = ((patients1['DEATHDATE'] - patients1['BIRTHDATE']).dt.days / 365.25).astype(int)
```

```python
print("\nShape (rows, columns):", patients1.shape)
```

```python
import pandas as pd
conditions1 = pd.read_csv('C:\\Users\\Mahi2\\capstone\\capstone_healthpredictors\\data\\condis.csv')
# View a few rows (table)
print(conditions1.head().to_string())
print("\nShape (rows, columns):", conditions1.shape)
```

```python
conditions1['memberid'] = conditions1['PATIENT'].str[:8]
conditions1.dtypes
```

```python
conditions1["START"] = pd.to_datetime(conditions1["START"], errors="coerce")
conditions1["STOP"] = pd.to_datetime(conditions1["STOP"], errors="coerce")
conditions1["CODE"] = conditions1["CODE"].astype("Int64").astype(str)
conditions1["CODE"] = conditions1["CODE"].fillna(0).astype("Int64").astype(str)
print("\nShape (rows, columns):", conditions1.shape)
```

```python
immunizations = pd.read_csv('C:\\Users\\Mahi2\\capstone\\capstone_healthpredictors\\data\\immunizations.csv')
# View a few rows (table)
print(immunizations.head().to_string())

print("\nShape (rows, columns):", immunizations.shape)
print(immunizations.columns.tolist())

immunizations["memberid"] = immunizations["PATIENT"].str[:8]
print(immunizations.columns.tolist())
```

```python
immunizations["DATE"] = pd.to_datetime(immunizations["DATE"], errors="coerce")
immunizations["CODE"] = immunizations["CODE"].astype("Int64").astype(str)
immunizations["CODE"] = ( immunizations["CODE"]
    .fillna(0)
    .astype("Int64")
    .astype(str)
)
immunizations["PATIENT"] = immunizations["PATIENT"].astype(str)
immunizations["memberid"] = immunizations["memberid"].astype(str)
```
